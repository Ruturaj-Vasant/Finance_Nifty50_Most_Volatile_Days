```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

print("Files in directory:", os.listdir())
```

```python
#df=pd.read_csv('NIFTY 500_Historical_PR_01011990to25032025.csv')
df=pd.read_csv('NIFTY 50_Historical_PR_01011990to25032025.csv')
df
```

```python
df = df.drop(columns=['Open','High','Low'])
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df = df.sort_values(by='Date', ascending=True).reset_index(drop=True)
type(df)
df
```

```python
df['Daily_Return']=df['Close'].pct_change()
df
```

```python
df['20DayVol'] = df['Daily_Return'].rolling(window=20).std()
df[15:25]
```

```python
df['Z20Vol']= (df["20DayVol"]-df['20DayVol'].rolling(window=250).mean().shift(1))/ df['20DayVol'].rolling(window=250).std().shift(1)
df[260:280]
```

```python
df = df.dropna(subset=['Z20Vol'])
df
```

```python
df = df.sort_values(by='Z20Vol', ascending=True)
df
```

```python
df.to_excel("NIFTY50_Zvol.xlsx", index=False)
```
