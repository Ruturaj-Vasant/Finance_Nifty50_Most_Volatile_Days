```python
import pandas as pd
import os

print("Files in directory:", os.listdir())
```

```python
df=pd.read_csv('NIFTY 50_Historical_PR_01011990to25032025.csv')
#df=pd.read_excel('assignment_VOL_data.xlsx')
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
df['Daily_returns']=df["Close"].pct_change()
df
```

```python
df['20-day_Volatility'] = df['Daily_returns'].rolling(window=20).std()

print(df.iloc[15:26])
```

```python
df["250_Mean"] = df["20-day_Volatility"].rolling(window=250).mean()
df["250_SD"] = df["20-day_Volatility"].rolling(window=250).std(ddof=0)  # Population SD

# Compute Zret20Vol using `250_Mean` instead of `.mean()` (to match Excel)
df["Z20Vol"] = (df["20-day_Volatility"] - df["250_Mean"].shift(1)) / df["250_SD"].shift(1)
output_file = "NIFTY50_Zret_Compare.xlsx"
df.to_excel(output_file, sheet_name="Compare_Zret", index=False)
df[260:280]
```

```python
df_filtered = df.dropna(subset=["Zret20Vol"])
df_sorted = df_filtered.sort_values(by="Zret20Vol", ascending=False)
df_sorted.head(-20)
output_file = "NIFTY50_Zret_Sorted.xlsx"
df_sorted.to_excel(output_file, sheet_name="Sorted_Zret", index=False)
```
